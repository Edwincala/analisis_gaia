## Base de datos

Se posee una base de datos de Gaia (DR3) (ESA) para la clasificación de estrellas, se tienen en cuenta los siguientes parámetros:
* RA_ICRS: Right ascension in the ICRS (International Celestial Reference System) coordinate system.
    (ESP: Ascensión recta en el sistema de coordenadas ICRS (Sistema Internacional de Referencia Celestial)).
    
* DE_ICRS: Declination in the ICRS coordinate system. (ESP: Declinación en el sistema de coordenadas ICRS.)
* Source: Unique identifier for an object in the Gaia catalogue. (ESP: Identificador único para un objeto en el catálogo Gaia)
* Plx: Trigonometric parallax of the celestial body in milliarcseconds. (ESP: Parallax trigonométrico del cuerpo celeste en miliarcsegundos.)
* PM: Total proper motion of the celestial body in the sky in milliarcseconds per year. (ESP: Movimiento propio total del cuerpo celeste en el cielo en milisegundos de arco por año.)
* pmRA: Proper motion component in the right ascension direction in milliarcseconds per year. (ESP: Componente del movimiento propio en la dirección de ascensión recta en milisegundos de arco por año.)
* pmDE: Proper motion component in the declination direction in milliarcseconds per year. (ESP: Componente del movimiento propio en la dirección de declinación en milisegundos de arco por año.)
* Gmag: Average apparent magnitude integrated in the G band. (ESP: Magnitud aparente media integrada en la banda G.)
* e_Gmag: Error in G magnitude. (ESP: Error en magnitud G.)
* BPmag: Average apparent magnitude integrated in the BP blue band. (ESP: Magnitud aparente promedio integrada en la banda azul de BP.)
* e_BPmag: Error associated with BP magnitude. (ESP: Error asociado con la magnitud de la PA.)
* RPmag: Average apparent magnitude integrated in the RP red band. (ESP: Magnitud aparente promedio integrada en la banda roja del RP.)
* e_RPmag: Error in RP magnitude. (ESP: Error en la magnitud RP.)
* GRVSmag: Average apparent magnitude integrated in the RVS band. (ESP: Magnitud aparente media integrada en la banda RVS.)
* e_GRVSmag: Error in RVS magnitude. (ESP: Error en la magnitud RVS.)
* BP-RP: BP-RP color index. (ESP: Índice de color BP-RP.)
* BP-G: BP-G color index. (ESP: Índice de color BP-G.)
* G-RP: G-RP color index. (ESP: Índice de color G-RP.)
* pscol: pseudocolor estimated by gaia in um<sup>-1</sup> (ESP: Pseudocolor estimado por Gaia en um<sup>-1</sup>)
* Teff: Estimated effective temperature of the celestial object by Gaia in Kelvins. (ESP: Temperatura efectiva estimada del objeto celeste por Gaia en Kelvins.)
* Dist: Distance to the celestial object: inverse of the parallax, in parsecs. (ESP: Distancia al objeto celeste: inversa del parallax, en pársecs.)
* Rad: Object radius estimate in terms of solar radius. (ESP: Estimación del radio del objeto en términos de radio solar.)
* Lum-Flame: Estimated object luminosity in terms of solar luminosity. (ESP: Luminosidad estimada del objeto en términos de luminosidad solar.)
* Mass-Flame: Mass estimate in terms of solar mass. (ESP: Estimación de masa en términos de masa solar.)
* Age-Flame: Celestial object age in giga years. (ESP: Edad de los objetos celestes en giga años. $1\text{ giga año} = 10^9\text{ años}$)
* z-Flame: Redshift in km/s. (ESP: Corrimiento al rojo en km/s.)
* SpType-ELS: Estimated spectral class by Gaia. (ESP: Clase espectral estimada por Gaia.)

Fuente: [kaggle datasets](https://www.kaggle.com/datasets/realkiller69/gaia-data-set-for-stellar-classificationdr3?resource=download)

#### Caracterización de las estrellas:

La metodología que usa ESA por medio de Gaia para el análisis de las estrellas vecinas comprende principalmente tres técnicas:

1. <strong>Astrometría:</strong> Medición de distancia, posición y movimiento propio de la estrella.
2. <strong>Fotometría:</strong> Observación de la edad, luminosidad, cambios en la luminosidad, temperatura, masa, color, polvo y metalicidad.
3. <strong>Espectrocopia:</strong> Determinación de la composición química, velocidad radial y de rotación por medio de los espectros de luz.

### Clasificación de las estrellas:

Según [Astrobitácora.com](https://www.astrobitacora.com/la-clasificacion-estelar/), existen tres factores principales para la clasificación de las estrellas, la clase espectral, la color y magnitud absoluta. La clase espectral es una especie de código que resume el estado de ionización de la estrella, y permite calcular la temperatura de la fotosfera y su densidad.

Existen 7 clases de estrellas definidas en el catálogo de Gaia: O, B, A, F, G, K y M.

* <strong>Clase O:</strong> Estas estrellas poseen una temperatura de entre 30000 y 60000 K, emiten un color azul. Debido a las altas temperaturas e intensidades de luminosidad, suelen poseer una vida corta, explotando de manera violenta, dando lugar a agujeros negros y estrellas de neutrones.

* <strong>Clase B:</strong> Poseen una temperatura de entre 10000 y 30000 K, la luz que emiten es de color azul también, suelen tener una masa de entre 2 y 16 veces la del Sol. No poseen corona (aura de plasma) ni zona de convección (zona en la que la energía es transmitida por convección).

* <strong>Clase A:</strong> Tienen una temperatura de entre 7500 y 10000 K. La luz que emiten es de color blanco, poseen una masa de entre 1.4 y 2.1 veces la del Sol. Se indica que los planetas masivos son formados alrededor de estrellas de este tipo.

* <strong>Clase F:</strong> La temperatura de estas estrellas oscila entre los 6000 y 7500 K, emiten un color blanco ligeramente amarillento, su masa es de entre 1 y 1.4 veces la masa del sol. Son consideradas estrellas enanas.

* <strong>Clase G:</strong> Su temperatura es de entre 5000 y 6000 K, tienen una masa de entre 0.8 y 1.2 veces la del Sol. Emiten luz de color blanco, el Sol pertenece a este grupo, y aunque vemos la luz de color amarillo, esto se debe a un fenómeno llamado la dispersión de Rayleigh. Su periodo de vida es de aproximadamente unos 10000 millones de años.

* <strong>Clase K:</strong> El rango de temperaturas es de entre 3500 y 5000 K, emiten luz de color anaranjado. Su masa es de entre 0.6 y 0.9 masas solares, poseen un periodo de vida de entre 15000 y 30000 millones de años. Debido a su baja transmisión de rayos ultravioleta, estas estrellas son objetivos para los astrónomos para buscar planetas con vida extraterrestre.

* <strong>Clase M: </strong> Estas estrellas poseen temperaturas de entre 2000 y 3500 K, son de un color rojizo, su etapa principal de evolución es del orden de billones de años, por lo que son bastante longevas a comparación de los otros tipos de estrellas.

## Generación del modelo de clasificación de estrellas.

### Separación del dataset para entrenamiento y testeo

Se utiliza la librería sklearn con su módulo train_test_split para la separación del dataset con un 30% dedicado al testeo del modelo y el 70% restante para su entrenamiento.

```python
from sklearn.model_selection import train_test_split
```

```python
training_data, test_data, train_class, test_class = train_test_split(
    df1.values,
    df["SpType-ELS"].values,
    test_size=0.3
)
```

Se implementa un modelo de árbol de decisión que tendrá en cuenta las posibles salidas para distintos intervalos de los valores numéricos presentes en la copia del dataframe generada y distintas combinaciones de los mismos para así predecir la posible clase de una estrella no identificada. El criterio usado es 'gini' que mide la impureza de los nodos del árbol de clasificación.

```python
from sklearn import tree
```

```python
arbol_decision = tree.DecisionTreeClassifier(criterion='gini', max_depth=8)

arbol = arbol_decision.fit(training_data, train_class)

accuracy = arbol_decision.score(test_data, test_class)

df1.head(5)
```

```python
analyzed_columns = list(df1.columns)
arbol_texto = tree.export_text(arbol, feature_names=analyzed_columns)

print(accuracy)
print(arbol_texto)
plt.figure(figsize=(60,24))
tree.plot_tree(arbol, feature_names=analyzed_columns)
plt.show()
```

## Evaluación del modelo

Es necesario revisar la precisión del algoritmo usado para la predicción o clasificación de las estrellas, la librería sklearn posee ya algunos parámetros de evaluación. Algunas de las métricas son: confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
def metricas(clases_reales, clases_predichas):
    matriz = confusion_matrix(clases_reales, clases_predichas)
    accuracy = accuracy_score(clases_reales, clases_predichas)
    precision = precision_score(clases_reales, clases_predichas)
    recall = recall_score(clases_reales, clases_predichas)
    f1 = f1_score(clases_reales, clases_predichas)
    return matriz, accuracy, precision, recall, f1
```

```python
def visualiza_metricas(clases_reales, clases_predichas, titulo):
    matriz, accuracy, precision, recall, f1 = metricas(clases_reales, clases_predichas)
    plt,figure(figsize=(3,3))
    matriz = pd.DataFrame(matriz, columns=["0: No aprobado", "1: Aprobado"])
    plt.matshow(matriz, cmap="Blues", vmin=0, vmax=10, fignum=1)
    plt.title("Predichas")
    plt.ylabel("reales")
    plt.xticks(range(len(matriz.columns)), matriz.columns, rotation=45)
    plt.yticks(range(len(matriz.columns)), matriz.columns)
    for i in range(len(matriz.columns)):
        for j in range(len(matriz.columns)):
            plt.text(i, j + 0.14, str(matriz.iloc[i,j]), fontsize=30, ha="center", va="center")
    plt.text(1.60, -0.30, titulo, fontsize=25, c="red")
    plt.text(2.1, 0.10, 'Exactitud: %0.2f' % accuracy, fontsize=20)
    plt.text(2.1, 0.40, 'Precisión: %0.2f' % precision, fontsize=20)
    plt.text(2.1, 0.70, 'Exhaustividad: %0.2f' % recall, fontsize=20)
    plt.text(2.1, 1.00, 'F1: %0.2f' % f1, fontsize=20)
    plt.show()
```
